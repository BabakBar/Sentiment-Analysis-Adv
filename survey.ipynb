```python
# Word cloud of survey results
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.colors import LinearSegmentedColormap
from wordcloud import WordCloud, STOPWORDS


survey = pd.read_csv("Survey.csv")
survey.info()

text_survey=str(survey)
text_survey
```

```python

colors = ["#BF0A30", "#002868"]
cmap = LinearSegmentedColormap.from_list("mycmap", colors)
my_stop_words = STOPWORDS.update(['Q10','Q9','the','ha','qu','think'])

#A function to generate the word cloud from text
def generate_basic_wordcloud(data):
    cloud = WordCloud(width=400,
                      height=330,
                      max_words=90,
                      background_color="white",
                      colormap=cmap,
                      stopwords=my_stop_words,
                      collocations=True).generate_from_text(text_survey)
    plt.figure(figsize=(10,8))
    plt.imshow(cloud, interpolation="bilinear")
    plt.axis('off')
    plt.show()
    
#Use the function to generate the word cloud
generate_basic_wordcloud(text_survey)
```
